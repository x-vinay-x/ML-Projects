```python
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x : round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
numeric_df = df_tracks.select_dtypes(include='number')
corr_df = numeric_df.corr(method="pearson")

# Plot the heatmap
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor='Black')
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# Create popularity group
df_tracks['popularity_group'] = pd.cut(df_tracks['popularity'], 
                                       bins=[0, 20, 40, 60, 80, 100], 
                                       labels=['0-20', '21-40', '41-60', '61-80', '81-100'])

# You can now do:
df_tracks['popularity_group'].value_counts().sort_index().plot(kind='bar')
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Prepare data: average duration per popularity group
total_dr = df_tracks.groupby('popularity_group')['duration'].mean().reset_index()

# Plot
fig_dims = (18,7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x='popularity_group', y='duration', data=total_dr, ax=ax, errwidth=0).set(title='Popularity vs Average Duration')
plt.xticks(rotation=90)
plt.ylabel('Average Duration')
plt.xlabel('Popularity Group')
plt.show()




```

```python
# Prepare data: average duration per popularity score
pop_duration = df_tracks.groupby('popularity', observed=True)['duration'].mean().reset_index()

# Plot
sns.set_style(style="whitegrid")
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize=fig_dims)

sns.lineplot(x='popularity', y='duration', data=pop_duration, ax=ax).set(title='Popularity vs Average Duration')

plt.xticks(rotation=60)
plt.ylabel('Average Duration')
plt.xlabel('Popularity')
plt.show()
```

```python
# Prepare data: average duration per popularity score (you can replace popularity with other columns too)
pop_duration = df_tracks.groupby('popularity', observed=True)['duration'].mean().reset_index()

# Set style
sns.set_style("whitegrid")
fig_dims = (15, 6)
fig, ax = plt.subplots(figsize=fig_dims)

# Use barplot with colorful palette (hsv is commonly used for rainbow style)
sns.barplot(x='popularity', y='duration', data=pop_duration, palette='hsv', ax=ax).set(title='Popularity vs Average Duration')

# Rotate x-axis labels
plt.xticks(rotation=60)
plt.ylabel('Average Duration')
plt.xlabel('Popularity')
plt.show()

```

```python
# Example with artist_name (if exists)
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")
sns.set_palette("rocket")

# Group by artist_name (top 20 artists by count)
top_artists = df_tracks['artist_name'].value_counts().head(20).index
df_top_artists = df_tracks[df_tracks['artist_name'].isin(top_artists)]

# Plot: Artist vs Average Duration
sns.barplot(y='artist_name', x='duration', data=df_top_artists, estimator=np.mean)
plt.title("Duration of Songs by Top 20 Artists")
plt.xlabel("Duration (seconds)")
plt.ylabel("Artist Name")
plt.show()
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize=(10,5))

# Top 10 artists by average popularity
famous = df_tracks.groupby('artist_name')['popularity'].mean().reset_index()
famous = famous.sort_values("popularity", ascending = False).head(10)

sns.barplot(y='artist_name', x='popularity', data = famous).set(title="Top 10 Artists by Popularity")
plt.show()
```
